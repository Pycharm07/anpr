```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
```

```python
img = cv2.imread('./Images/car_3.jpg')
```

```python
def plot_images(img1, img2, title1= '', title2 = ''):
	fig = plt.figure(figsize = [15, 15])
	ax1 = fig.add_subplot(121)
	ax1.imshow(img1, cmap = "gray")
	ax1.set(xticks = [], yticks=[], title = title1)

	ax1 = fig.add_subplot(122)
	ax1.imshow(img2, cmap = "gray")
	ax1.set(xticks = [], yticks=[], title = title2)	
	plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blurred_img = cv2.GaussianBlur(img, (21, 21), 0)
```

```python
plot_images(gray, blurred_img, title1 = 'Gray', title2 = 'Blur')
```

```python
mask = np.zeros((512, 512, 3), dtype=np.uint8)
```

```python
print(type(mask))
```

```python
mask = cv2.circle(mask, (120, 120), 50, (255,0,0), 2)
```

```python
plot_images(mask, mask, title1 = 'Gray', title2 = 'Blur')
```
